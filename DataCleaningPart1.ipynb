```python
df=pd.read_csv('titanic.csv',header=0,sep=";")
```

```python
df
```

```python
# drop rows with nan
df.dropna(axis=0,how='all') # equivalent to df.dropna(how="all")
```

```python
df.shape
```

```python
# Evaluate cells in the dataFrame for null : dataframe.isnull()
df.isnull()

# Is there any row which is null completely? syntax : dataframe.isnull().all()
# .all() queries row or columns  
print(df.isnull().all(axis=1))
# These two are same
print(df.isnull().all(axis=0))
print(df.isnull().all())

# Number of rows which are completely null
print("Number of null tuples",df.isnull().all(axis=1).sum())

#index of line/lines which is completely null
df.index[df.isnull().all(axis=1)].tolist()
```
